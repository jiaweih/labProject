###Maurer - Livneh(14)

```python

from mpl_toolkits.basemap import Basemap, cm
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
#pylab inline --no-import-all

vars = {'Prec','Tmax', 'Tmin'}
seasons = ['Spring','Summer', 'Autumn', 'Winter']
store_dir = '/raid3/jiawei/homero/plots/L14/'
source_dir = '/raid3/jiawei/homero/data/maurer/L14/' 

for var in vars:
    #define the template for the filename
    ftemplate = source_dir+'maurer'+'.{}.nc' #######################
    # Create figure   
    fig = plt.figure(figsize=(20,20))
    # Layout panels in figure
    grid = gridspec.GridSpec(4, 2, height_ratios=[1,1,1,1,0.05], wspace=0.05, hspace=0.05)
    idx = 0
    for season in seasons:
#        plt.subplot(index)
        # open panel
        ax = fig.add_subplot(grid[idx])
        # read netcdf file
        nc = Dataset(ftemplate.format(season), 'r')
        info = nc.variables[var]
        data=info[0][:]
        lon=nc.variables['lon'][:]
        lat=nc.variables['lat'][:]
        nc.close()
        # plot
        m = Basemap(projection='merc',llcrnrlat=25.25,urcrnrlat=52.85,\
        llcrnrlon=-124.75,urcrnrlon=-67.25,lat_ts=30,resolution='c')
        m.drawcoastlines(linewidth=0.25)
        xi, yi = m(*np.meshgrid(lon, lat))
        cs = m.contourf(xi,yi,data,levels = np.linspace(-4.5,4.5,10),cmap=plt.cm.seismic, extend="both")
        #plt.title(season, fontsize=15)
        # add text
        ax.text(0.10, 0.97, seasons[idx], verticalalignment='top', horizontalalignment='left',
                transform=ax.transAxes, color='black', fontsize=16)
        ax.text(0.97, 0.97, var, verticalalignment='top', horizontalalignment='right',
                transform=ax.transAxes, color='black', fontsize=16)
        ax.text(0.5, 0.97, "1950-2000", verticalalignment='top', horizontalalignment='center',
                transform=ax.transAxes, color='black', fontsize=16) 
        ax.text(0.5, 0.91, "Maurer-Livneh(14)", verticalalignment='top', horizontalalignment='center',
                transform=ax.transAxes, color='black', fontsize=16)
        idx += 1    
    idx = 0
    for season in seasons:
#        plt.subplot(index)
        # open panel
        ax = fig.add_subplot(grid[idx+4])
        # read netcdf file
        nc = Dataset(ftemplate.format(season), 'r')
        info = nc.variables[var]
        data=info[0][:]
        lon=nc.variables['lon'][:]
        lat=nc.variables['lat'][:]
        nc.close()
        # plot
        m = Basemap(projection='merc',llcrnrlat=48.25,urcrnrlat=55.25,\
        llcrnrlon=-124.75,urcrnrlon=-109.25,lat_ts=45,resolution='c')
        m.drawcoastlines(linewidth=0.75)
        #m.drawmapboundary(fill_color='aqua')
        
        xi, yi = m(*np.meshgrid(lon, lat))
        cs = m.contourf(xi,yi,data,levels = np.linspace(-4.5,4.5,10),cmap=plt.cm.seismic, extend="both")
        #plt.title(season, fontsize=15)
        # add text
        ax.text(0.10, 0.97, seasons[idx], verticalalignment='top', horizontalalignment='left',
                transform=ax.transAxes, color='black', fontsize=16)
        ax.text(0.97, 0.97, var, verticalalignment='top', horizontalalignment='right',
                transform=ax.transAxes, color='black', fontsize=16)
        ax.text(0.5, 0.97, "1950-2000", verticalalignment='top', horizontalalignment='center',
                transform=ax.transAxes, color='black', fontsize=16) 
        ax.text(0.5, 0.91, "Maurer-Livneh(14)", verticalalignment='top', horizontalalignment='center',
                transform=ax.transAxes, color='black', fontsize=16)
        idx += 1  
    # get axes bounds.
    pos = ax.get_position()
    l, b, w, h = pos.bounds 
    # create axes instance for colorbar on bottom.
    cax = plt.axes([l-.2, b-0.03, w, 0.01])
    # draw colorbar on bottom.
    cbar = plt.colorbar(cax=cax,orientation='horizontal')
    cbar.ax.tick_params(labelsize=10)
    if var == 'Prec':
        cbar.set_label(' (mm/day)')
    else:
        cbar.set_label(' (C)')
    plt.show() 
        

    fig.savefig(store_dir+'Maurer_'+var+'_1950_2000.png')
```
