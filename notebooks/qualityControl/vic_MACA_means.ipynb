```python
from mpl_toolkits.basemap import Basemap, cm
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
#pylab inline --no-import-all


names = ['Evap','G','H','L','Precip','Qair','Qsb','Qs','Rnet','RTEMP','SHIn','SoilMst1','SoilMst2','SoilMst3','SWE','WDEW']
store_dir = '/home/raid3/jiawei/vic_MACA/data/netCDF/means/figures/'
source_dir = '/home/raid3/jiawei/vic_MACA/data/netCDF/means/'

#define the template for the filename
ftemplate = source_dir+'hist.sh'+'.{}.nc' #######################
fig = plt.figure(figsize=(30,30))
grid = gridspec.GridSpec(4, 4, height_ratios=[1,1,1,1,0.05], wspace=0.05, hspace=0.10)
idx = 0

for ss in grid:
    var = names[idx]
    nc = Dataset(ftemplate.format(var), 'r')
    info = nc.variables[var]
    data = info[0][:]
    lon = nc.variables['lon'][:]
    lat = nc.variables['lat'][:]
    nc.close()
    
    ax = fig.add_subplot(ss)  
    m = Basemap(projection='merc',llcrnrlat=41.2188,urcrnrlat=52.8438,\
            llcrnrlon=-124.594,urcrnrlon=-109.781,lat_ts=30,resolution='c')
    m.drawcoastlines(linewidth=0.25)
    xi, yi = m(*np.meshgrid(lon, lat))
    cs = m.contourf(xi,yi,data,cmap=plt.cm.seismic,)
    
    # add text
    ax.text(0.97, 0.97, var, verticalalignment='top', horizontalalignment='right',
                transform=ax.transAxes, color='black', fontsize=18)
    cbar = m.colorbar(cs,location='bottom',pad='4%')
    cbar.ax.tick_params(labelsize=13)              
    idx += 1

fig.savefig(store_dir+'mean.png')
```
